## Interacting with an API

It is a type of software interface, offering a service to other pieces of software. A document or standard that describes how to build such a connection or interface is called an API specification. ### A brief introduction to `requests`

`curl` is a small but very powerful application that can be used to transfer data from a remote server using HTTP/S, a [request-response](https://en.wikipedia.org/wiki/Request%E2%80%93response) protocol.

We could use `curl` to download the data from `gov.uk` and make it available to our notebook, but there is a better way to do that natively in Python with `requests`.

`requests` is a powerful Python library that can be used to:
  - send a _request_ to a remote server
  - receive, decode and store the _response_ in a variable

![title](img/request.png)

`requests` on PyPI: https://pypi.org/project/requests/

### `requests` in practice

Send the request and save the response in a variable (`bank_holidays_response`).

````python
import requests

bank_holidays_response = requests.get('https://www.gov.uk/bank-holidays.json')
````

At this point we may want to check that everything went well and we got the data.

````python
if bank_holidays_response.status_code == 200:
    print("I got the data!")
else:
    print(f"Something went wrong")
````

Let's see what is inside `bank_holidays_response`. Every response will contain a `status_code` (a numerical value that tell us if the request was successful or not) and other things.

`bank_holidays_response.headers` contain additional data related to our request.

We can access what is inside a response using `text` (`bank_holidays_response.text`) or even `content` (`bank_holidays_response.content`), but in this case we already know that the response is in JSON:

````python
bank_holidays_response.headers["Content-Type"]
````

So we can use the method `.json()` to access the JSON content of the response directly as a Python dictionary.

````python
type(bank_holidays_response.json())
````

````python
bank_holidays_response.json()
````

And now we can assign it to a variable and start exploring:

````python
data = bank_holidays_response.json()
````

````python
data.keys()
````

````python
data["england-and-wales"].keys()
````

````python
data["england-and-wales"]["events"]
````

And finally we can extract all the bank holidays for a specific year:

````python
bank_holidays_2026 = [event for event in data["england-and-wales"]["events"] if "2026" in event["date"]]
````

````python
# show all the bank holidays

bank_holidays_2026
````

````python
# how many bank holidays in 2026?

len(bank_holidays_2026)
````

````python
# show only the name and the date for each bank holiday

for item in bank_holidays_2026:
    print(f"Bank holiday: {item['title']}, date: {item['date']}")
````

## 2 - Weather forecast

In this section we are going to use the APIs from Open-Meteo (https://open-meteo.com/en/docs).

They are well documented, reliable and don't need authentication.

Open-Meteo also provides a useful preview of the data and an API builder.

We will try to get a 7-days forecast for a specific location and plot a diagram of `temperature_2m`.

The base URL is:
```
https://api.open-meteo.com/v1/forecast?latitude=LAT&longitude=LON&hourly=VARIABLES
```

We will need to replace:
- `LAT` with the latitude
- `LON` with the longitude
- `VARIABLES` with a weather variable or two or more weather variables separated by a comma (`,`)

We can use the Maidenhead Library a location. If you search for the library on Google Maps ([link](https://maps.app.goo.gl/kCnBFgmkGPdWM93C9)) and then right-click on the pin you should be able to get the coordinates:
`51.5222087244349, -0.7179850346446116`

Let's start by storing this information in some variables. 

````python
lat = "51.52"
lon = "-0.71"
variables = "temperature_2m"

open_meteo_URL = f"https://api.open-meteo.com/v1/forecast?latitude={lat}&longitude={lon}&hourly={variables}"
````

Now we can use requests and get the data from Open-Meteo.

````python
import requests

open_meteo_response = requests.get(open_meteo_URL)
````

(note that we don't need to import `requests` again if the previous import was executed and the kernel was not restarted)

Check that everything went well and we got the data.

````python
open_meteo_response.status_code
````

What is the format of the data? Let's check the headers. And if the `Content-Type` is `application/json` we can use the method `.json()`.

````python
open_meteo_response.headers
````

````python
data = open_meteo_response.json()
````

Let's start exploring the data, print the `keys()`.

````python
data.keys()
````

````python
data["hourly"].keys()
````

````python
data["hourly"]
````

### A brief introduction to `pandas`

`pandas` is one of the most important libraries used to perform data analysis with Python.

We will use `pandas`, `seaborn` and `matplotlib` to extract information from our data and plot a diagram.

Some resources to get started with `pandas`:
- https://pandas.pydata.org/docs/user_guide/10min.html
- https://www.datacamp.com/tutorial/pandas

````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
````

Create a new _dataframe_ which contains the data (time, temperature value).

````python
df = pd.DataFrame(data["hourly"])
````

````python
df
````

`info()` will tell us how `pandas` is storing the data internally.

````python
df.info()
````

We can convert the datetime, which is a `str` at the moment, to a proper format (`datetime64`).

````python
df["time"] = pd.to_datetime(df['time'])
````

````python
df.info()
````

````python
df.describe()
````

Plot a diagram using `seaborn`.

````python
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='time', y='temperature_2m')

# 4. Customize plot
plt.title('Hourly Temperature (forecast)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
````
