## Interacting with an API

> An **application programming interface** (**API**) is a connection between computers or between computer programs. It is a type of software interface, offering a service to other pieces of software. A document or standard that describes how to build such a connection or interface is called an API specification. A computer system that meets this standard is said to implement or expose an API. (...)
> 
> [API - Wikipedia](https://en.wikipedia.org/wiki/API)

APIs are often used to exchange data between applications or, for example, control/trigger actions.

JSON is often used when interacting with APIs as it's _human-readable_ but can also be used by applications.

> **JSON** (**JavaScript Object Notation**, pronounced /ˈdʒeɪsən/ or /ˈdʒeɪˌsɒn/) is an open standard file format and data interchange format that uses human-readable text to store and transmit data objects consisting of name–value pairs and arrays (or other serializable values). It is a commonly used data format with diverse uses in electronic data interchange, including that of web applications with servers. (...)
> 
> [JSON - Wikipedia](https://en.wikipedia.org/wiki/JSON)

To handle data in JSON format with Python we can use the JSON module from the standard library (https://docs.python.org/3/library/json.html)

### A brief introduction to `requests`

`curl` is a small but very powerful application that can be used to transfer data from a remote server using HTTP/S, a [request-response](https://en.wikipedia.org/wiki/Request%E2%80%93response) protocol.

We could use `curl` to download the data from `gov.uk` and make it available to our notebook, but there is a better way to do that natively in Python with `requests`.

`requests` is a powerful Python library that can be used to:
  - send a _request_ to a remote server
  - receive, decode and store the _response_ in a variable

![title](img/request.png)

`requests` on PyPI: https://pypi.org/project/requests/

Send the request and save the response in a variable (`bank_holidays_response`).

````python
import requests

bank_holidays_response = requests.get('https://www.gov.uk/bank-holidays.json')
````

At this point we may want to check that everything went well and we got the data.

````python
if bank_holidays_response.status_code == 200:
    print("I got the data!")
else:
    print(f"Something went wrong")
````

Let's see what is inside `bank_holidays_response`. Every response will contain a `status_code` (a numerical value that tell us if the request was successful or not) and other things.

`bank_holidays_response.headers` contain additional data related to our request.

We can access what is inside a response using `text` (`bank_holidays_response.text`) or even `content` (`bank_holidays_response.content`), but in this case we already know that the response is in JSON:

````python
bank_holidays_response.headers["Content-Type"]
````

So we can use the method `.json()` to access the JSON content of the response directly as a Python dictionary.

````python
type(bank_holidays_response.json())
````

````python
bank_holidays_response.json()
````

And now we can assign it to a variable and start exploring:

````python
data = bank_holidays_response.json()
````

````python
data.keys()
````

````python
data["england-and-wales"].keys()
````

````python
data["england-and-wales"]["events"]
````

And finally we can extract all the bank holidays for a specific year:

````python
bank_holidays_2026 = [event for event in data["england-and-wales"]["events"] if "2026" in event["date"]]
````

````python
# show all the bank holidays

bank_holidays_2026
````

````python
# how many bank holidays in 2026?

len(bank_holidays_2026)
````

````python
# show only the name and the date for each bank holiday

for item in bank_holidays_2026:
    print(f"Bank holiday: {item['title']}, date: {item['date']}")
````
