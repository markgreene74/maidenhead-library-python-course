## What are virtual environments?

Virtual environments, as the name suggests, provide separate "spaces" where packages can be installed to satisfy the requirements of a specific application without interfering with the "system" Python (Linux and other *NIX OS) or with other applications.

They are built on a "base" version of Python and are considered "disposable" (they don't need to be committed to VCS or copied from one system to another). With the right tools can provide reproducible and "easy to rebuild" environments to build or run applications on many different systems.

One of the main tools used to create virtual environments is the `venv` module.

## Some practice

Virtual environments and the tools connected are difficult to explore in a notebook, especially as this is probably already running in a virtual environment or on Binder (which understandably has some limitations about what we can run or install).

How do we know if we are in a virtual environment? See this [answer on Stack Overflow](https://stackoverflow.com/a/1883251) or [this message on the Python forum](https://discuss.python.org/t/determining-if-code-is-running-in-a-venv/26862/3).

```python
import sys
print(f"{sys.prefix=}")
print(f"{sys.base_prefix=}")
```

```python
message = "This is likely running in a venv." if sys.prefix != sys.base_prefix else "This is likely not running in a venv."
print(message)
```
