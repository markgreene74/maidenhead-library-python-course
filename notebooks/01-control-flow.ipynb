## Flow control tools

The main tools to control the flow in Python are:
- `if` \[`elif`] `else` statements
- `for` loops
- `while` loops
- `match` statements

They have different purposes, let's see some examples.

## `if` \[`elif`] `else` statements

An `if` \[`elif`] `else` statement is used to evaluate a boolean condition (`True`/`False`) and run a block of code if it is `True`. The `elif` statement is used to evaluate multiple conditions after the first one, and the `else` statement is reached only if all the previous conditions are false. Both `elif` and `else` are optional.

The most basic syntax of an `if` block is:

```python
if condition:
    print("condition is True")
```

This syntax can be extended with an `else`:

```python
if condition:
    print("condition is True")
else:
    print("condition is False")
```

And we can even add `elif` to test more conditions:

```python
my_favourite_fruit = "apple"  # replace apple with your favourite fruit

if my_favourite_fruit == "apple":
    print("You like apples!")
elif my_favourite_fruit == "banana":
    print("You like bananas!")
elif my_favourite_fruit == "mango":
    print("You like mangos!")
else:
    print("You don't like bananas, apples or mangos ¯\_(ツ)_/¯")
```

The basic syntax of a `for` loop is:

```python

    # do something here
```

## `while` loops

A `while` loop is used to execute a block of code again and again, as long as the condition being evaluated is `True`. If the condition never changes (is always `True`) the `while` loop will run forever ... (or until we do something to exit it or stop it).

The basic syntax of a `while` loop is:

```python
while condition:
    # do something here
```

This `while` loop will run forever:

```python
while True:
    print("hello")
```

This `while` loop will stop at some point:

```python
from random import randint

while True
    r = randint(0,10)
    print(f"The value of r is: {r}")
    if r == 10:
        break
```

In this last example we are using a simple `if` and the keyword `break` to exit the loop.

## `match` statements

A `match` statement to compare the result of an expression against multiple values. You can think of a `match` statement as a more complex (and efficient) `if` \[`elif`] `else` statement.

The syntax of a `match` statement is:

```python
match expression:
    case value_1:
        print("This is value_1")
    case value_2:
        print("This is value_2")
    case value_3:
        print("This is value_3")
    case _:
        print("No value was matched")
```

The `_` in this case acts as a _wildcard_ that catches any possible value of the expression, and in this case works as an `else` (the last code block is executed if there is no match).

The `match` statement is a relatively new addition to Python (introduced in [3.10](https://docs.python.org/3/whatsnew/3.10.html)) and it may not be there if you are using an older version of the interpreter.

A practical example:

```python
my_favourite_fruit = "apple"  # replace apple with your favourite fruit

match my_favourite_fruit:
    case "apple":
        print("You like apples!")
    case "banana":
        print("You like bananas!")
    case "mango":
        print("You like mangos!")
    case _:
        print("You don't like bananas, apples or mangos ¯\_(ツ)_/¯")
```

## Functions

Another very important way to control the flow of our program is based on _functions_.

A function, in very simple terms, is a block of code that can be re-used. By using functions you can make your code more efficient, avoid repetition and reduce the possibility of creating an error (a bug).

A function in Python is commonly defined with the keyword `def` followed by the function name and a few other parts (parenthesis, arguments, colon), some of which are optional (arguments).

```python
def my_function():
    print("Hello! This is my function")
```

Once it is defined, the function can be called with `my_function()`. In this example the function:
- does not take any argument
- does not return anything (implicitly returns `None`)

Let's see another example:

```python
def say_hi(name):
    print(f"Hi {name}!")
```

In this example the function takes one argument (a variable called `name`), which is used to print a message, but again does not return anything. Remember that when calling this function you will need to specify a value for the argument `name`.

How about functions that do ... something and return a value? Let's see another example:

```python
def minutes_to_seconds(minutes):
    print(f"Converting {minutes} minutes to seconds")
    result = minutes * 60

    return result
```

This function will not print the result, but instead it will `return` it, and then we can do what we want with it (print it, use it with another function...).

For example:

```python
print(minutes_to_seconds(1))
```

Or even better:

```python
one_minute_in_seconds = minutes_to_seconds(1)
print(f"One minute is {one_minute_in_seconds} seconds")
print(f"Half a minute is {one_minute_in_seconds / 2} seconds")
```
